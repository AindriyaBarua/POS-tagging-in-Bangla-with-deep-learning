```python
from __future__ import print_function
import keras
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop, Adam
from sklearn.model_selection import GridSearchCV
from keras.models import Sequential
from keras.wrappers.scikit_learn import KerasClassifier
import numpy as np
import gensim, logging
from gensim.models import word2vec
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from numpy.random import randint
from numpy import argmax
from keras.utils.np_utils import to_categorical
```

```python
with open("filtered.txt","r") as readfile:
    types = [line.split("_") for line in readfile]
```

```python
X=[]
y=[]
c=0
#print (types)
for i in range(len(types)):
    X.append(types[i][0])
    y.append(types[i][1])

label = [k.strip("\n") for k in y]
#print(label)
X = [X]
```

```python
label = [k.strip("\n") for k in y]
```

```python
model = gensim.models.Word2Vec(X, size=500, window=5, min_count=1, workers=4) #input shape=100
```

```python
m,n = np.shape(X)
dataset = []
for i in range(n):
    dataset.append(model.wv[X[0][i]])
```

```python
le = preprocessing.LabelEncoder()

tags = le.fit(label)
le.fit(label)

tags=le.transform(label)
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset, tags, test_size=0.20, random_state=42)
```

```python
import numpy as np
print(np.shape(X_train), np.shape(X_test), np.shape(y_train), np.shape(y_test))
```

```python
num_classes = len(set(label))
```

```python
print(np.shape(X_train), np.shape(X_test), np.shape(y_train), np.shape(y_test))
```

```python
X_train = np.stack(X_train, axis = 0)
print(np.shape(X_train))
X_test = np.stack(X_test, axis = 0)
print(np.shape(X_test))
```

```python
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
model = Sequential()
model.add(Dense(16, activation='sigmoid', input_shape=(500,))) # changed here for changing size arror
#model.add(Dense(8, activation='sigmoid'))
#model.add(Dropout(0.2))
model.add(Dense(num_classes, activation='sigmoid'))#changed from sigmoid(92%)
```

```python
batch_size = 32
epochs = 500
model.summary()

model.compile(loss='categorical_crossentropy',
              optimizer=Adam(lr=0.1),
              metrics=['accuracy'])

model.fit(X_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_split = 0.1)
```
